```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from datetime import datetime
```

```python
df=pd.read_csv("NFLX.csv")
df
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.body()
```

```python
sns.lineplot(x=df.index,y=df["Volume"],label="Volume")
plt.title("Volume of stock vs time")
plt.show()
```

Netflix stock price:HIGH,OPEN AND CLOSE

```python
df.plot(y=["High","Open","Close"],title="Netflix stock Price",ylabel="Price")
plt.show()
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))# ploting 3 subplot in a single plot
df.groupby(df.index.day).mean().plot(y="Volume",ylabel="Volume",ax=ax1,xlabel="Day")
df.groupby(df.index.month).mean().plot(y="Volume",ylabel="Volume",ax=ax2,xlabel="Month",color="red")
df.groupby(df.index.year).mean().plot(y="Volume",ylabel="Volume",ax=ax3,xlabel="Year",color="green")
plt.show()
```

WITHOUT MEAN CALCULATION

```python
fig, (ax1,ax2,ax3)=plt.subplots(3,figsize=(25,40))# ploting 3 subplot in a single plot
df.groupby(df.index.day).plot(y="Volume",ylabel="Volume",ax=ax1,xlabel="Day")
df.groupby(df.index.month).plot(y="Volume",ylabel="Volume",ax=ax2,xlabel="Month")
df.groupby(df.index.year).plot(y="Volume",ylabel="Volume",ax=ax3,xlabel="Year")
plt.show()
```
